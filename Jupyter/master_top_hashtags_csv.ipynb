```python
import pandas as pd
import os
import datetime

YEAR = 2022

DATA_FOLDER_PATH = '..\\data\\'

BACKUP_RAW_CSV_FILES_FOLDER_PATH = DATA_FOLDER_PATH + "archive_backup\\" # up Jupyter\ Folder, into data\archive_backup\
RAW_CSV_FILES_FOLDER_PATH = DATA_FOLDER_PATH + "archive\\" # up Jupyter\ Folder, into data\archive\
TOP_HASHTAGS_PER_DAY_FOLDER_PATH =  DATA_FOLDER_PATH + "top_hashtags_per_day\\" # up Jupyter\ Folder, into data\top_hashtags_per_day\

is_first_csv = True
for filename in os.listdir(TOP_HASHTAGS_PER_DAY_FOLDER_PATH):
    if filename[0] == 'm':
        continue
    
    print("Combining " + filename)
    if is_first_csv:
        df0 = pd.read_csv(os.path.join(TOP_HASHTAGS_PER_DAY_FOLDER_PATH, filename))
        df0['daily_rank'] = df0.index + 1
        df0['date'] = datetime.date(int(filename[0:4]), int(filename[5:7]), int(filename[8:10]))
        is_first_csv = False     
    else:
        dfn = pd.read_csv(os.path.join(TOP_HASHTAGS_PER_DAY_FOLDER_PATH, filename))
        dfn['daily_rank'] = dfn.index + 1
        dfn['date'] = datetime.date(int(filename[0:4]), int(filename[5:7]), int(filename[8:10]))
        df0 = pd.concat([df0, dfn], ignore_index=True)
       
df0.to_csv(os.path.join(TOP_HASHTAGS_PER_DAY_FOLDER_PATH, "master_top_hashtags_per_day_with_ranking.csv"), index=False)
```
